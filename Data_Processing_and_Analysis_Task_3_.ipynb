**Problem Statement 1: Employee Performance Bonus Eligibility**


```python
def get_top_performers(employees):
    if not employees:
        return [], 0
    max_score = max(employees.values())
    winners = [name for name, score in employees.items() if score == max_score]
    return winners, max_score
if __name__ == "__main__":
    employees = {
        "Ravi": 92,
        "Anita": 88,
        "Kiran": 92,
        "Suresh": 85
    }
    top_names, top_score = get_top_performers(employees)
    print(f"Top Performers Eligible for Bonus: {', '.join(top_names)} (Score: {top_score})")
```

**Problem Statement 2: Search Query Keyword Analysis**

```python
def get_repeated_keywords(query):
    query = query.lower()
    for char in ".,!?;:":
        query = query.replace(char, "")
    words = query.split()
    counts = {}
    for word in words:
        counts[word] = counts.get(word, 0) + 1
    return {word: count for word, count in counts.items() if count > 1}
if __name__ == "__main__":
    text = "Buy mobile phone buy phone online"
    repeated_words = get_repeated_keywords(text)
    print(repeated_words)
```

**Problem Statement 3: Sensor Data Validation**

```python
def validate_sensor_data(readings):
    valid_readings = []
    for hour, value in enumerate(readings):
        if value % 2 == 0:
            valid_readings.append((hour, value))
            return valid_readings
if __name__ == "__main__":
    sensor_readings = [3, 4, 7, 8, 10, 12, 5]
    results = validate_sensor_data(sensor_readings)
    print("Valid Sensor Readings (Hour, Value):")
    print(results)
```

**Problem Statement 4: Email Domain Usage Analysis**

```python
def analyze_domains(emails):
    if not emails:
        return {}
    counts = {}
    total = len(emails)
    for email in emails:
        domain = email.split('@')[1]
        counts[domain] = counts.get(domain, 0) + 1
    analysis = {}
    for domain, count in counts.items():
        percentage = (count / total) * 100
        analysis[domain] = f"{int(percentage)}%"
    return analysis
if __name__ == "__main__":
    email_list = [
        "ravi@gmail.com",
        "anita@yahoo.com",
        "kiran@gmail.com",
        "suresh@gmail.com",
        "meena@yahoo.com"
    ]
    results = analyze_domains(email_list)
    for domain, percent in results.items():
        print(f"{domain}: {percent}")
```

**Problem Statement 5: Sales Spike Detection**

```python
def detect_sales_spikes(sales):
    if not sales:
        return
    avg = sum(sales) / len(sales)
    for day, value in enumerate(sales, 1):
        increase = (value - avg) / avg
        if increase >= 0.29:
            print(f"Day {day}: {value}")
if __name__ == "__main__":
    sales_data = [1200, 1500, 900, 2200, 1400, 3000]
    detect_sales_spikes(sales_data)
```

**Problem Statement 6: Duplicate User ID Detection**

```python
def find_duplicate_ids(user_ids):
    counts = {}
    for uid in user_ids:
        counts[uid] = counts.get(uid, 0) + 1
    for uid, count in counts.items():
        if count > 1:
            print(f"{uid} → {count} times")
if __name__ == "__main__":
    user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
    find_duplicate_ids(user_ids)
```
