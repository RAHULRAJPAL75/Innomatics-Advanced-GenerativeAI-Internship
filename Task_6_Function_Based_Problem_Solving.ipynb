**Problem Statement 1:Smart Parking Lot Management System**

```python
def manage_parking_lot(capacity, logs):
    current_parked = 0
    peak_usage = 0
    for action in logs:
        if action.upper() == "IN":
            current_parked += 1
        elif action.upper() == "OUT":
            if current_parked > 0:
                current_parked -= 1
        if current_parked > peak_usage:
            peak_usage = current_parked
    if current_parked >= capacity:
        status = "Parking Full / Alert"
    else:
        status = "Available"
    return current_parked, status, peak_usage
if __name__ == "__main__":
    parking_capacity = 50
    vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
    current, status, peak = manage_parking_lot(parking_capacity, vehicle_logs)
    print(f"Currently Parked Vehicles: {current}")
    print(f"Parking Status: {status}")
    print(f"Peak Occupancy Recorded: {peak}")
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**


```python
def analyze_theatre_occupancy(total_seats, booked_seats_list):
    booked_count = len(booked_seats_list)
    occupancy_pct = (booked_count / total_seats) * 100
    if occupancy_pct == 100:
        status = "Housefull"
        suggestion = "Urgent: Open additional show immediately."
    elif occupancy_pct >= 75:
        status = "Almost Full"
        suggestion = "Recommended: Schedule an extra show for this movie."
    elif occupancy_pct >= 50:
        status = "Moderate Occupancy"
        suggestion = "Status: Sufficient seating available."
    else:
        status = "Low Occupancy"
        suggestion = "Status: No additional shows needed."
    return occupancy_pct, status, suggestion
if __name__ == "__main__":
    TOTAL = 200
    BOOKED_DATA = [1] * 150
    pct, status, advice = analyze_theatre_occupancy(TOTAL, BOOKED_DATA)
    print(f"Occupancy: {int(pct)}%")
    print(f"Show Status: {status}")
    print(f"Management Suggestion: {advice}")
```

**Problem Statement 4:Cloud Server Load Classification System**


```python
def classify_server_load(cpu_readings):
    total_load = 0
    for reading in cpu_readings:
        total_load += reading
    num_readings = len(cpu_readings)
    avg_load = total_load / num_readings
    if avg_load < 50:
        status = "Normal"
    elif 50 <= avg_load <= 80:
        status = "Warning"
    else:
        status = "Critical"
    return avg_load, status
if __name__ == "__main__":
    cpu_data = [45, 60, 70, 85, 90]
    avg, status = classify_server_load(cpu_data)
    print(f"Average CPU Load: {int(avg)}%")
    print(f"Server Status: {status}")
```

**Problem Statement 5:Smart Classroom Resource Usage Monitor**

```python
def monitor_classroom_resources(usage_dict):
    OVERUSE_LIMIT = 8
    overused_list = []
    for resource, hours in usage_dict.items():
        if hours > OVERUSE_LIMIT:
            overused_list.append(resource)
    energy_alert = "Yes" if len(overused_list) > 0 else "No"
    overused_str = ", ".join(overused_list) if overused_list else "None"
    print(f"Overused Resources: {overused_str}")
    print(f"Energy Alert: {energy_alert}")
if __name__ == "__main__":
    classroom_data = {
        "Projector": 6,
        "AC": 9,
        "Lights": 4
    }
    monitor_classroom_resources(classroom_data)
```

 **Problem Statement 6: Online Event Registration Capacity Controller**

```python
def manage_event_registration(capacity, total_registrations):
    if total_registrations <= capacity:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
if __name__ == "__main__":
    EVENT_CAPACITY = 100
    TOTAL_SIGNUPS = 105
    manage_event_registration(EVENT_CAPACITY, TOTAL_SIGNUPS)
```
