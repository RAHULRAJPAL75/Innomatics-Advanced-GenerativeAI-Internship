**Problem 1: Social Media - Post Engagement Analyzer**

```python
def analyze_post_engagement(daily_likes):
    if not daily_likes:
        return 0, "No Engagement Data"
    total_likes = 0
    for like_count in daily_likes:
        total_likes += like_count
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    return total_likes, status
if __name__ == "__main__":
    post_data = [150, 300, 200, 400]
    total, result = analyze_post_engagement(post_data)
    print(f"Total Likes: {total}")
    print(f"Post Status: {result}")
```

**Problem 2: Healthcare - Medicine Stock Alert System**

**Problem 3: Agriculture - Rainfall Adequacy Checker**

```python
def check_rainfall_adequacy(rainfall_data, required_level):
    total_rainfall = 0
    for record in rainfall_data:
        total_rainfall += record
    average_rainfall = total_rainfall / len(rainfall_data)
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")
    print("-" * 30)
if __name__ == "__main__":
    seasonal_data = [65, 80, 71]
    threshold = 70
    check_rainfall_adequacy(seasonal_data, threshold)
```

**Problem 4: Social Media - Duplicate Account Detection**

```python
def detect_duplicate_accounts(usernames):
    unique_usernames = set(usernames)
    if len(unique_usernames) < len(usernames):
        status = "Yes"
    else:
        status = "No"
    print(f"Duplicate Accounts Found: {status}")
if __name__ == "__main__":
    user_list_1 = ["user_alpha", "dev_spark", "user_alpha", "admin_01"]
    print("Testing List 1:")
    detect_duplicate_accounts(user_list_1)
    user_list_2 = ["pixel_pioneer", "cloud_walker", "data_wiz"]
    print("\nTesting List 2:")
    detect_duplicate_accounts(user_list_2)
```

**Problem 6: Agriculture - Premium Crop Price Filter**

```python
def filter_premium_crops(price_list):
    premium_threshold = 2000
    premium_crops = []
    for price in price_list:
        if price > premium_threshold:
            premium_crops.append(price)
    print(f"Premium Crops: {premium_crops}")
if __name__ == "__main__":
    market_prices = [1500, 2500, 1800, 3200, 2000]
    filter_premium_crops(market_prices)
```

**Problem 8: Banking - Daily Transaction Limit Checker**

```python
def validate_transaction(amount):
    DAILY_LIMIT = 50000
    if amount <= DAILY_LIMIT:
        status = "Approved"
    else:
        status = "Rejected"
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")
    print("-" * 30)
if __name__ == "__main__":
    validate_transaction(60000)
    validate_transaction(45000)
    validate_transaction(50000)
```

**Problem 9: E-Learning - Student Attendance Eligibility System**

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    for status in attendance_list:
        if status.upper() == "P":  #
            present_days += 1
    attendance_percentage = (present_days / total_days) * 100
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")
    print("-" * 30)
if __name__ == "__main__":
    student_record_1 = ["P", "P", "P", "P", "A", "P", "P", "P", "A", "P"]
    check_exam_eligibility(student_record_1)
    student_record_2 = ["P", "A", "P", "A", "P", "P", "A", "P", "A", "P"]
    check_exam_eligibility(student_record_2)
```

**Problem 10: Smart Electricity Bill Analyzer**


```python
def calculate_electricity_bill(units):
    bill_amount = 0.0
    if units <= 100:
        bill_amount = units * 3
    elif units <= 200:
        bill_amount = (100 * 3) + ((units - 100) * 5)
    else:
        bill_amount = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill_amount < 500:
        usage_status = "Low Usage"
    elif 500 <= bill_amount <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    return bill_amount, usage_status
if __name__ == "__main__":
    print("--- Smart Meter Billing Dashboard ---")
    try:
        units_consumed = float(input("Enter units consumed: "))
        total_bill, status = calculate_electricity_bill(units_consumed)
        print("-" * 35)
        print(f"Units Consumed: {units_consumed}")
        print(f"Total Bill: ₹{total_bill:.2f}")
        print(f"Usage Status: {status}")
        print("-" * 35)
    except ValueError:
        print("Invalid input! Please enter a numeric value for units.")
```
