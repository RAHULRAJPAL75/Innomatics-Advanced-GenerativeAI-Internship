**Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))
            if amount < 50:
                print("Error: Minimum recharge amount is ₹50. Please try again.")
                continue
            if amount not in valid_plans:
                print(f"Error: ₹{amount} is not a valid plan. Valid plans are: {valid_plans}")
                continue
            print(f"Recharge of ₹{amount} successful!")
            break
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
if __name__ == "__main__":
    validate_recharge()
```

**Problem 2: Inventory Reorder Alert System**

```python
def check_stock_levels(stock_data):
    for item, quantity in stock_data.items():
        if quantity < 15:
            print(f"{item}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{item}: Stock OK (Stock = {quantity})")
if __name__ == "__main__":
    orbital_inventory = {
        "Oxygen Tanks": 42,
        "Fuel Cells": 12,
        "Medical Kits": 8,
        "Water Recyclers": 20,
        "Protein Packs": 10,
        "Carbon Scrubbers": 25,
        "Thermal Blankets": 5,
        "Repair Bots": 30
    }
    check_stock_levels(orbital_inventory)
```

**Problem 3: Student Result Processing System**

```python
def evaluate_performance(scores):
    if not scores:
        return 0, "No Data"
    total_points = 0
    for point in scores:
        total_points += point
    avg_score = total_points / len(scores)
    status = "Pass" if avg_score >= 50 else "Fail"
    return avg_score, status
if __name__ == "__main__":
    marks_list = [98,40,69,32,95]
    average, result = evaluate_performance(marks_list)
    print(f"Average Score: {average:.2f}")
    print(f"Final Status: {result}")
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, is_peak):
    BASE_FARE = 50.0
    RATE_PER_KM = 12.0
    PEAK_SURGE_MULTIPLIER = 1.25
    subtotal = BASE_FARE + (distance * RATE_PER_KM)
    return subtotal * PEAK_SURGE_MULTIPLIER if is_peak else subtotal
def run_fare_estimator():
    print("=== Smart Cab Fare Estimator ===")
    while True:
        try:
            km_input = input("\nEnter trip distance (km): ").strip()
            km = float(km_input)
            if km < 0:
                print("Error: Distance cannot be negative.")
                continue
            peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
            is_peak = peak_input in ['yes', 'y']
            total_fare = calculate_fare(km, is_peak)
            print("-" * 30)
            print(f"Distance: {km} km")
            print(f"Peak Hour: {'Yes (+25%)' if is_peak else 'No'}")
            print(f"Total Estimate: ₹{total_fare:.2f}")
            print("-" * 30)
            if input("Calculate another trip? (y/n): ").strip().lower() != 'y':
                print("Safe travels! Goodbye.")
                break
        except ValueError:
            print("Error: Please enter a valid number for the distance.")
if __name__ == "__main__":
    run_fare_estimator()
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance_eligibility(attendance_record):
    if not attendance_record:
        return 0, 0, "No Data"
    present_days = 0
    for status in attendance_record:
        if status.upper() == "P":
            present_days += 1
    total_days = len(attendance_record)
    attendance_pct = (present_days / total_days) * 100
    status = "Eligible" if attendance_pct >= 75 else "Not Eligible"
    return attendance_pct, status, present_days
if __name__ == "__main__":
    log = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "A"]
    percentage, result, present_count = check_attendance_eligibility(log)
    print(f"Total Days Present: {present_count}")
    print(f"Attendance Percentage: {percentage}%")
    print(f"Eligibility Status: {result}")
```

**Problem 6: Password Strength Checker**

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."
    special_chars = set("@#$")
    has_digit = False
    has_special = False
    for char in password:
        if char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True
        if has_digit and has_special:
            break
    if has_digit and has_special:
        return "Strong: Security requirements met."
    reasons = []
    if not has_digit:
        reasons.append("at least one digit (0-9)")
    if not has_special:
        reasons.append("one special character (@, #, or $)")
    return f"Weak: Missing {' and '.join(reasons)}."
if __name__ == "__main__":
    print("--- Security Portal: Password Validation ---")
    user_input = input("Create a new password: ").strip()
    result = check_password_strength(user_input)
    print(f"\nResult: {result}")
```
