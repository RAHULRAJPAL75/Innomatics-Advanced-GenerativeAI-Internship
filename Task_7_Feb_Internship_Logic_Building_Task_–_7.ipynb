**Problem 1: Smart Light Controller**

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False
    def set_status(self, action):
        if action.upper() == "ON":
            self.is_on = True
        elif action.upper() == "OFF":
            self.is_on = False
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")
my_light = SmartLight("Bedroom Light")
my_light.set_status("ON")
```

**Problem 2: Employee ID Card System**

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
emp1 = Employee("Rahul", "EMP102", "IT")
emp1.display_id_card()
```

**Problem 3: Mobile Contact Record**


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
contact1 = Contact("Anita", "9876543210")
contact1.display_contact()
```

**Problem 4: Product Price Tag Generator**

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def generate_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
item1 = Product("Headphones", 2499)
item1.generate_tag()
```

**Problem 5: Movie Rating Display System**

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
movie1 = Movie("Inception", 4.8)
movie1.display_details()
```

**Problem 6: Delivery Address Manager**

```python
class DeliveryManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")
shipment1 = DeliveryManager("Suman", "Hyderabad")
shipment1.print_details()
```
